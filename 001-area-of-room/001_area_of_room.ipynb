### Programme - Area of a room

```python
#defining the area function:
def area(width, length):
 try:
  width = float(width)
  length = float(length)
  area = width * length
  return area
 except ValueError:
  return "Invalid input! width and length must be numbers."
 
# prompt user for dimensions:
width_input = input("Enter the width of a room: ")
length_input = input("Enter the length of a room: ")

# calculate and display the area:
result = area(width_input, length_input)
print(f"Total area: {result} square metres")
```
