### Programme: Bottle Deposit

```python
# prompt user for the bottles returned:
sml_bottles = input("How many bottles did you return less than or equal to 1 liter? ")
lrg_bottles = input("How many bottles did you return greater than 1 liter? ")

# exception handling for invalid input:
try:
 # convert the input to integers:
 sml_bottles = int(sml_bottles)
 lrg_bottles = int(lrg_bottles)

 # calculate the total refund:
 sml_bottle_refund = sml_bottles * 0.10
 lrg_bottle_refund = lrg_bottles * 0.25
 total_refund = round(sml_bottle_refund + lrg_bottle_refund, 2)
 print(f"Your total refund is: ${total_refund}")

except ValueError:
 print("Error! Invalid input.")
```
